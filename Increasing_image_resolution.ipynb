```python
import cv2
from cv2 import dnn_superres
```

```python
super_res = dnn_superres.DnnSuperResImpl_create()
```

```python
model = 'C:\\Users\\Qadeer\\OneDrive\\Documents\\projects\\esrgan_project\\models\\LapSRN_x8.pb'
super_res.readModel(model)
```

```python
super_res.setModel('lapsrn', 8)
```

```python
super_res.setPreferableBackend(cv2.dnn.DNN_BACKEND_CUDA)
super_res.setPreferableTarget(cv2.dnn.DNN_TARGET_CUDA)
```

```python
img = cv2.imread('C:\\Users\\Qadeer\\OneDrive\\Documents\\projects\\esrgan_project\\media\\download (1).jpg')
```

```python
import matplotlib.pyplot as plt
import cv2
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB if using OpenCV
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
upscaled = super_res.upsample(img)
```

```python
plt.imshow(cv2.cvtColor(upscaled, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB if using OpenCV
plt.title('Upscaled Image')
plt.axis('off')
plt.show()
```

```python
cv2.imwrite('C:\\Users\\Qadeer\\OneDrive\\Documents\\projects\\esrgan_project\\media\\download(upscaled).jpg', upscaled)
```
